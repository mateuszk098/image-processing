```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
frame = Image.open("./data/00-puppy.jpg")
img = np.array(frame)
plt.imshow(img)
plt.show()
```

```python
# Red channel values 0 to 255 (0 is no red and 255 is full red).
red_channel = img[..., 0]
plt.imshow(red_channel, cmap="gray")  # Closer to white means more red.
plt.show()
```

```python
# Green channel values.
green_channel = img[..., 1]
plt.imshow(green_channel, cmap="gray")
plt.show()
```

```python
# Blue channel values.
blue_channel = img[..., 2]
plt.imshow(blue_channel, cmap="gray")  # Ears are almost black, so there is no blue.
plt.show()
```

```python
# Zero out the red channel.
no_red = img.copy()
no_red[..., 0] = 0
plt.imshow(no_red)
plt.show()
```
